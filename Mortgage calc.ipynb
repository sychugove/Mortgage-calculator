```python
import pandas as pd
import pylab as plt
pd.set_option('display.float_format', lambda x: '%.0F' % x)
```

```python
def monthly_payment(loan, percentage, period):
    month_perc = percentage/12
    monthly_payment = loan * (month_perc * (1+ month_perc)**(period*12))/((1+ month_perc)**(period*12)-1)
    return(monthly_payment)  
```

```python
def plotting(df_low_period):
    plt.plot(range(df_low_period.shape[0]), df_low_period.debt, label = 'Проценты')
    plt.plot(range(df_low_period.shape[0]), df_low_period.mainloan, label = 'Основной платеж')
    #plt.plot(range(df_low_period.shape[0]), df_low_period.shape[0] * [df_low_period.mainloan.max()], 'r--')
    #plt.plot(range(df_low_period.shape[0]), df_low_period.shape[0] * [df_low_period.debt.min()], 'g--')
    plt.title('Проценты и основной долг')
    plt.legend()
    plt.show()
```

```python
period = 20
percentage = 0.05
loan = 7000000
payment = 56500
bov = [31892, 481193, 204200, 157000]
bov_month = [240, 239, 238, 232]

bov_df = pd.DataFrame(bov, index=bov_month, columns = ['bigoverpay'])
report_num = 0
mortgage_report = pd.DataFrame(columns=['year', 'month', 'm_payment', 'percent', 'main_l', 'overpay', 'big_overpay',
                                        'realpay_on_debt', 'payment' ,'loan'])

for i in range(period , 0 , -1):
    for j in range(i*12 , (i-1)*12 , -1):
        if j in bov_month:
            m = bov_df.loc[j].values[0]
        else:
            m = 0
        mp = monthly_payment(loan, percentage, j/12)
        percent = percentage/12 * loan
        main_l = mp - percent
        overpay = payment - mp
        real_pay_on_debt = main_l + overpay + m
        #print(i, j, monthly_payment(loan, percentage, j/12), percent, main_l, loan)
        mortgage_report.loc[report_num] = [i, j, monthly_payment(loan, percentage, j/12), percent, main_l, overpay, m,
                                           real_pay_on_debt, payment + m, loan]
        report_num += 1
        loan -= real_pay_on_debt    

mortgage_report['payment_downgrade'] = mortgage_report['m_payment'] - mortgage_report['m_payment'].shift()

col = list(mortgage_report.columns)
col = col[:3] + col[-1:]  + col[4:-1]
mortgage_report = mortgage_report[col]

mortgage_report.head(26)
```
